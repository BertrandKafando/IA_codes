```python
import numpy as np 
import matplotlib.pyplot as plt
import numpy.linalg as lag
```

```python
#data
x=np.array([[2,2],[-1,0],[3,0],[5,3]])
Y=np.array([-1,0,1,2])
X=np.hstack((x,np.ones((4,1))))
```

```python
np.shape(x)
```

```python
np.shape(Y)
```

```python
X
```

```python
x.shape
```

```python
beta=lag.inv(X.T.dot(X)).dot(X.T).dot(Y)
beta
```

```python
 def predfonct(X,beta):
        return X.dot(beta)
        
```

```python
predfonct(X,beta)
```

```python
plt.scatter(x[:,1],predfonct(X,beta))
#plt.plot(x[:,0],Y,"r")
ax = plt.axes(projection = '3d')
ax.scatter3D(x[:,0],x[:,1],predfonct(X,beta))
#ax2 = plt.axes(projection = '3d')
ax.scatter3D(x[:,0],x[:,1],Y,"r")
```

```python
reg=Y-predfonct(X,beta)
Reg=reg.T.dot(reg)
ste=Y-Y.mean()
STE=ste.T.dot(ste)
Reg
```

```python
def coef_deter(y,pred):
    u=((y-pred)**2).sum()
    v=((y-y.mean())**2).sum()
    return 1-u/v
```

```python
coef_deter(Y,predfonct(X,beta))
```
